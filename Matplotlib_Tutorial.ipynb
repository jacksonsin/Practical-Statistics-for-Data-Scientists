**Import Libraries**

```python
# Magic command for IPython
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Basic Graph**

```python
x = [0, 1,2,3,4]
y = [0, 2,4,6,8]

# Resize your graph (dpi specifies pixels per inch, when saving recommended to use 300 for high resolution)
plt.figure(figsize=(5,3), dpi=200)

# Line 1
# Keyword argument notation
#plt.plot(x,y, color='red',linewidth='2', linestyle='--', marker='.', markersize='10', markeredgecolor='blue', label='2x')

# Shorthand notation
# fmt = [color][marker][line]
plt.plot(x,y,'r.--',label='2x')

# Line 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2, 'b', label='x^2')
plt.plot(x2[5:],x2[5:]**2, 'b--')

# Add a title (specifies font parameters with fontdict)
plt.title('First Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})

# X and Y labels
plt.xlabel('X axis')
plt.ylabel('Y axis')

# X, Y axis tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8])

# Add a legend
plt.legend()

# Save figure (dpi specifies pixels per inch, when saving recommended to use 300 for high resolution)
plt.savefig('graph.png', dpi=200)

# Show plot
plt.show() # To remove the object text [<matplotlib.lines.Line2D at 0x7fe1781e6890>]
```
