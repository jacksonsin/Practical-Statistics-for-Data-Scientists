```python
# experiment
entitlement_d = []
entitlement = []
value = []
test = data["Entitlement Details"]

columns = ["Entitlement Details","Entitlement","Value"]
for row in [data["Entitlement Details"]]:
    for z in range(len(row)):
        for i in [row[z][:-1]]:
            x = i.split(";")
        for j in x:
            entitlement += list(j.split(' - ')[-2:-1])
            value += list(j.split(' - ')[-1:])
            entitlement_d += [test[z]]

    values = [entitlement_d,entitlement, value]
    zipped = zip(columns,values)
    a_dictionary = dict(zipped)

export_file = pd.DataFrame.from_dict(a_dictionary)
```
